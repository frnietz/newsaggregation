```python
pytrend = TrendReq(hl='tr-TR', tz=120)
```

```python
trends = pytrend.related_queries()
```

```python
stopwords = set(STOPWORDS) 
#stopwords.add("borsa")
```

```python
wc = WordCloud(
    width = 1000, 
    height = 600,    
    background_color = 'white',
    colormap = 'rainbow',
    stopwords = stopwords,
    max_words = 500,
    max_font_size = 150,
    min_font_size = 15,
    contour_width = 2, 
    contour_color = 'dodgerblue'
)
```

```python
df_trending_searches_tr = pytrend.trending_searches(pn='turkey')
df_today_searches_tr = pytrend.today_searches(pn='TR')
```

```python
text2 = ' '.join(df_trending_searches_tr[0].to_list())
wc.generate_from_text(text2)
fig2, axes= plt.subplots(figsize=(10,5),clear=True)
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

```python
text3 = ' '.join(df_today_searches_tr.to_list())
wc.generate_from_text(text3)
fig3, axes= plt.subplots(figsize=(10,5),clear=True)
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
